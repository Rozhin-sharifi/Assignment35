```python
import numpy as np
import pandas as pd
import random
import sklearn
import matplotlib.pyplot as plt
from scipy import stats
```

```python
N = 100

X = []
Y = []

for i in range(N):
  X.append(np.random.randint(60, 240, dtype=int))
  y = random.uniform(0.25, 0.5) * X[i]
  Y.append(y)

plt.scatter(X, Y)
```

```python
# m = (X.XT)^-1 *YXT
X = np.array(X)
Y = np.array(Y)

X = X.reshape(100, 1)
Y = Y.reshape(100, 1)

print(X.shape, Y.shape)
```

```python
m1 = random.uniform(0.6, 0.8)
m2 = random.uniform(0.1, 0.3)

Y_pred1 = X * m1
Y_pred2 = X * m2

plt.scatter(X, Y)

plt.plot(X, Y_pred1, c= 'red')
plt.plot(X, Y_pred2, c= 'red')
plt.xlabel('Height')
plt.ylabel('Weight')
```
